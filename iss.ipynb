```python
import requests
import time
import matplotlib.pyplot as plt

data_points = []
num_points = 100
interval = 10

for _ in range(num_points):
    response = requests.get("http://api.open-notify.org/iss-now.json")
    data = response.json()
    
    latitude = float(data["iss_position"]["latitude"])
    longitude = float(data["iss_position"]["longitude"])
    timestamp = data["timestamp"]
    
    data_points.append((latitude, longitude, timestamp))
    
    time.sleep(interval)

print("Data points generated:", len(data_points))

latitudes = [data[0] for data in data_points]
longitudes = [data[1] for data in data_points]
timestamps = [data[2] for data in data_points]

plt.figure(figsize=(10, 6))
plt.scatter(longitudes, latitudes, c=timestamps, cmap='cool', alpha=0.6)
plt.colorbar(label='Timestamp')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Path of the International Space Station')
plt.grid(True)
plt.show()

```
